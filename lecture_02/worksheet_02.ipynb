### Effective Programming

a) What is a drawback of the top down approach?

In the top-down approach, coders think about code in a higher-level. Rather than think strictly about the specifics of how each function will work, in the top-down approach, pseudocode is generally created that explains the overview of how the program will work and how each function relates to another. Here, functions can be declared but implemented later, allowing you to still think about them as you plan your entire program. One drawback of this approach is that you can not run code until you have implemented most of the functions when planning out the higher-level pseudo code. 

b) What is a drawback of the bottom up approach?

In the Bottom up Approach, one drawback is that it takes a lot more planning when coding, and sometimes, you can end up creating functions that end up not being used in the end. This is because this approach usually requires users to implement low level functions first, followed by higher-order functions. Thus, if a higher-level function did not actually need an implemented lower-level function, time may be wasted. However, one good thing about this approach is that it allows users to test from the get-to. 


c) What are 3 things you can do to have a better debugging experience?

There are many things you can do to have a better debugging experience. Firstly, you should try your best to stay organized by splitting up the different parts of your code into functions. This makes unit testing a lot easier and can help you identify where exactly a bug is quicker. In addition, if a program is large, we should try our best to compartmentalize, whether that be splitting up our code in multiple files or folders. This allows us to understand the structure of our code in a larger depth, making debugging easier. Lastly, we should always think about the side effects of our code and what assumptions we make when we write it. For both our sake and those that read our code, creating comments regarding these topics and just understanding them will point you in the right direction when you need to debug.

a) Complete the code below to read in a filepath to the `train.csv` and returns the DataFrame.

```python
import pandas as pd

df = pd.read_csv('train.csv')
df.describe()
```

b) Complete the code so it returns the number of rows that have at least one empty column value

```python
print("there are " +  str(df.isnull().any(axis=1).sum()) + " rows with at least one empty value")
```

c) Complete the code below to remove all columns with more than 200 NaN values

```python
df = df.drop(df.columns[df.apply(lambda col: col.isnull().sum() > 200)], axis=1)
df.columns
```

d) Complete the code below to replaces `male` with 0 and `female` with 1

```python
df['Sex'] = df['Sex'].replace({'male' : 0, 'female': 1})
df.head()
```

e) Complete the code below to add four columns `First Name`, `Middle Name`, `Last Name`, and `Title` corresponding to the value in the `name` column.

For example: `Braund, Mr. Owen Harris` would be: 

|First Name | Middle Name | Last Name | Title |
|-----------|-------------|-----------|-------|
| Owen      |  Harris     |  Braund   | Mr    |

Anything not clearly one of the above 4 categories can be ignored.

```python
df[['First Name', 'Middle Name', 'Last Name', 'Title']] = df['Name'].str.split(' ', expand=True)[[2, 3, 0, 1]].apply(lambda x: x.str.rstrip('.')).apply(lambda x: x.str.rstrip(','))

df.head(5)
```

f) Complete the code below to replace all missing ages with the average age

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df.head()
```
